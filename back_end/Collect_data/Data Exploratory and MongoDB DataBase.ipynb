```python
def todict():
    import pandas as pd
    df = pd.read_csv("school_shooting.csv")
    df['Date'] = pd.to_datetime(df['Date'])
    column_names = list(df.columns)
    final_list = []
    name_list = []
    for item in column_names:
        name_list.append(item)
        final_list.append([i for i in df[item]])
        
    dictionary = dict(zip(name_list, final_list))
    
    return dictionary 

    
    
```
